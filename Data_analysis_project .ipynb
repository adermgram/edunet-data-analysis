```python
# 19. Which is the most sold bike here?
most_sold_bike = two_wheelers['Car_Name'].value_counts().idxmax()
most_sold_bike_count = two_wheelers['Car_Name'].value_counts().max()
print("Most sold bike:", most_sold_bike)
print("Count:", most_sold_bike_count)
```

```python
# 20. Deals in bikes exceeding general expectations

# Calculating the average price of bikes
average_bike_price = two_wheelers['Selling_Price'].mean()

# Filter for bikes sold above the average price
high_value_bikes = two_wheelers[two_wheelers['Selling_Price'] > average_bike_price]

print("Deals exceeding general expectations:")
print(high_value_bikes.head())
```

```python
# 21. Can we find out data of only cars from this data ?
cars = df[df["Present_Price"] >= 3.5]
cars.head()
```

```python
# Present_Price vs Kms_Driven

from matplotlib import pyplot as plt
cars.plot(kind='scatter', x='Present_Price', y='Kms_Driven', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Selling_Price

from matplotlib import pyplot as plt
cars['Selling_Price'].plot(kind='hist', bins=20, title='Selling_Price')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# title Seller_Type vs Transmission

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(4, 4))
df_2dhist = pd.DataFrame({
    x_label: grp['Transmission'].value_counts()
    for x_label, grp in cars.groupby('Seller_Type')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Seller_Type')
_ = plt.ylabel('Transmission')
```
